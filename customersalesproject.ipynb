```python
import pandas as pd

df = pd.read_csv("C:\projects\customer sales\data\customer_sales_data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Gender'] = df['Gender'].fillna('Unknown')
```

```python
df['Customer_Rating'] = df['Customer_Rating'].fillna(df['Customer_Rating'].mean())
```

```python
df.dropna(subset=['Product_Name', 'Unit_Price', 'Quantity'], inplace=True)
```

```python
df.to_csv("cleaned_customer_sales.csv", index=False)
```
